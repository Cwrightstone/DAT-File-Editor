```python
# Check the contents of your DAT file below. 
data = open('2022 DTM MI Annual - TEST.txt', 'rt')
file = data.read()
data.close()
#print(file)

```

```python
# Empty list for all the data to be entered into the headers
ROWCode = []
Milepost = []
LocationDescription = []
PermanentComments = []
# InspectionDate = [], its in the pattern TBD
ACP = []
On = []
Off = []
InspectionRemarks = []
Technician = []
InspectionLatitude = []
InspectionLongitude = []
Latitude = []
Longitude = []
```

```python
# Start finding the data you wish to extract to your tabluar and add them to their respected empty list

#RC = re.findall("") = Why is the DEMO poping as b7755de8-7ce4-4765-851f-6c534087ce83? Is this the code for the import?

MP = re.findall("Milepost�(.*)��C", file)
for m in MP:
    Milepost.append(MP)
#print(Milepost)

LD = re.findall("Location Description�(.*)��C", file)
for l in LD:
    LocationDescription.append(LD)
#print(LD)

#PC = re.findall("Permanent Comments�(.*)��C", file)
#for p in PC:
    #PermanentComments.append(PC)
#print(PC)

IR = re.findall("Inspection Remarks�(.*)��C", file)
for i in IR:
    InspectionRemarks.append(IR)
#print(IR)

AC = re.findall("AC�(.*)��A2", file)
for a in AC:
    ACP.append(AC)
#print(AC)

ON = re.findall("On�(.*)��V2", file)
for o in ON:
    On.append(ON)
#print(ON)

OFF = re.findall("Off�(.*)��V2", file)
for f in OFF:
    Off.append(OFF)
#print(OFF)

T = re.findall("Technician�(.*)��C", file)
for t in T:
    Technician.append(T)
#print(T)

IL = re.findall("Inspection Latitude�(.*)��GLAT", file)
for il in IL:
    InspectionLatitude.append(IL)
#print(IL)

ILon = re.findall("Inspection Longitude�(.*)��GLONG", file)
for ilon in ILon:
    InspectionLongitude.append(ILon)
#print(ILon)

Lat = re.findall("Latitude�(.*)��GLAT", file)
for lat in Lat:
    Latitude.append(Lat)
#print(Lat)

Lon = re.findall("Longitude�(.*)��GLONG", file)
for lon in Lon:
    Longitude.append(Lon)
#print(Lon)
```
